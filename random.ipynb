```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0.5  # Probability of success in Bernoulli trials
N_values = [5, 10, 25, 100, 1000, 10000]  # Different sequence lengths
num_sequences = 10000  # Number of sequences to generate

# Set up the subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Iterate over different N values
for i, N in enumerate(N_values):
    # Generate sequences of Bernoulli random variables
    sequences = np.random.binomial(1, mu, (num_sequences, N))
    
    # Compute the mean m for each sequence
    means = np.mean(sequences, axis=1)
    
    # Plot the histogram of m
    ax = axs[i//3, i%3]  # Select subplot
    ax.hist(means, bins=30, density=True, alpha=0.7, color='blue')
    ax.set_title(f'N = {N}')
    ax.set_xlabel('Mean (m)')
    ax.set_ylabel('Density')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
